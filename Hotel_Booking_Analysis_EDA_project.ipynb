````python
# Chart - 7 visualization code
# Finding Number of cancelation

df_copy['is_canceled'].value_counts()
````

````python
# Finding cancelation by repeated guests
df_copy[['is_canceled','is_repeated_guest']].value_counts().unstack()
````

````python
# Creating function for showing the percentage in countplot
def countplotpercent(df_copy,xvariable,ax):

    # 1. how many X categories
    ##   check for NaN and remove
    numX=len([x for x in df[xvariable].unique() if x==x])

    # 2. The bars are created in hue order, organize them
    bars = ax.patches
    ## 2a. For each X variable
    for ind in range(numX):
        ## 2b. Get every hue bar
        ##     ex. 8 X categories, 4 hues =>
        ##    [0, 8, 16, 24] are hue bars for 1st X category
        hueBars=bars[ind:][::numX]
        ## 2c. Get the total height (for percentages)
        total = sum([x.get_height() for x in hueBars])

        # 3. Print the percentage on the bars
        for bar in hueBars:
            ax.text(bar.get_x() + bar.get_width()/2.,
                    bar.get_height(),
                    f'{bar.get_height()/total:.0%}',
                    ha="center",va="bottom")
     
````

````python
# Visulalization of repeated guest cancelation
plt.figure(figsize=(10,5))
ax=sns.countplot (x= 'is_canceled', data= df_copy, hue= 'is_repeated_guest')
plt.xticks(size=15)
plt.xlabel('Not Canceled(0) or Canceled(1)',size=15)
plt.yticks(size=15)
plt.ylabel('count',size=15)
plt.title("Repeated guest relation with cancelation")
countplotpercent(df_copy,'is_canceled',ax)
     
````
